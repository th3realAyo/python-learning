```python
# The bool() function allows you to evaluate any value,
# and give you True or False in return

print(bool("Hello"))
print(bool(16))
print(bool(0))
print(bool(0.0))
print(bool(None))
print(bool(False))
print(bool([]))
print(bool({}))
print(bool())
print(bool(["apple", "cherry", "banana"]))

class myclass():
  def __len__(self):
    return 0

myobj = myclass()
print(bool(myobj))


# Functions can Return a Boolean
def myFunction() :
  return True

print(myFunction())

def myFunction():
  return True

if myFunction():
  print('Yes')
else:
  print('No')
```

```python
is_student = True
has_id = False

# Create new boolean from existing ones
can_enter = is_student and has_id  # True if BOTH are True
print(can_enter)  # False

can_get_discount = is_student or has_id  # True if AT LEAST ONE is True
print(can_get_discount)  # True

is_visitor = not is_student  # Opposite of is_student
print(is_visitor)  # False
```
