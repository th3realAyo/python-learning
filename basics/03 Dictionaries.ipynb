```python
students = {
    "Ayo": [80, 90, 75],
    "Bola": [60, 65, 70],
    "Tunde": [88, 79, 93]
}


for student, score in students.items():
    print(f"Name: {student} : Score: {score}")

print()

for name, score in students.items():
    print(f"Name: {name} : Avg score: {sum(score) / len(score):.2f}")

students['Shola'] = [76, 87, 90]
students['Ayo'] = [79, 89, 76]

print()
for name, score in students.items():
    average_score = round(sum(score) / len(score), 1)
    if average_score > 80:
        print(f"Name: {name} : Avg Score: {average_score}")
```
