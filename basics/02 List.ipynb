**LIST COMPREHENSION**
# syntax = [expression for value in iterable if condintion]

```python
doubles = [x * 2 for x in range(1, 11)]
triples = [y * 3 for y in range(1, 11)]
squares = [z * z for z in range(1, 11)]


print(doubles)
print(triples)
print(squares)


fruits = ['apple', 'banana', 'coconut']
fruits = [fruit.upper() for fruit in fruits]
print(fruits)


#With conditions

numbers = [ 1, 5, 7, -2, 0, -2, 19]

postive = [num for num in numbers if num >= 0]
print(postive)
```

```python
# LIST TRAVERSAL
horsemen = ['war', 'famine', 'pestilence', 'death']

i = 0
while i < 4:
    print(horsemen[i])
    i += 1
```

```python
# LIST LENGHT
horsemen = ["war", "famine", "pestilence", "death"]
i = 0

while i < len(horsemen):
    print(horsemen[i])
    i = i + 1
```
