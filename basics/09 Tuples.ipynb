```python
import random

for i, num in enumerate(range(5)):
    x = random.random()
    print(round(x * 2), )

high = 5
for i in range(high):
    new_high = random.random() * high
    print(new_high)
```

```python
def randList(n):
    s = [0] * n
    for i in range(n):
        s[i] = random.random()
    return s
print(randList(5))
```
